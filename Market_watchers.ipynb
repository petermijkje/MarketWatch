```python
import pandas_datareader
import datetime
```

```python
import pandas_datareader.data as web
from matplotlib import pyplot as plt
```

```python
start = datetime.datetime(2012,1,1)
end = datetime.datetime(2021,1,1)
```

```python
tesla = web.DataReader('TSLA','yahoo',start,end)
```

```python
tesla.head()
```

```python
ford = web.DataReader('F','yahoo',start,end)
gm = web.DataReader('GM','yahoo',start,end)
```

```python
ford.head()
```

```python
gm.head()
```

```python
tesla['Open'].plot(label='Tesla',figsize=(16,8),title='opening prices')
gm['Open'].plot(label='GM')
ford['Open'].plot(label='Ford')
plt.legend()
```

```python
tesla['Volume'].plot(label='Tesla',figsize=(16,8),title='opening prices')
gm['Volume'].plot(label='GM',figsize=(16,8),title='opening prices')
ford['Volume'].plot(label='Ford',figsize=(16,8),title='opening prices')
plt.legend()
```
